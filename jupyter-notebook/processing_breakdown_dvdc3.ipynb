```python
import os
import sys
import subprocess
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from pyarrow import fs
import pyarrow.parquet as pq
import pandas as pd
import numpy as np

from tqdm import tqdm
os.environ["HADOOP_CONF_DIR"] = "/path/to/hadoop/conf/"
os.environ["JAVA_HOME"] = "/path/to/java"
os.environ["HADOOP_HOME"] = "/path/to/hadoop"
os.environ["ARROW_LIBHDFS_DIR"] = "/path/to/hadoop/lib/"
os.environ["CLASSPATH"] = subprocess.check_output(
    "$HADOOP_HOME/bin/hadoop classpath --glob", shell=True
).decode("utf-8")

hdfs = fs.HadoopFileSystem(
    host="hdfs://your-hdfs-host", port=8020
)
```

```python
path_to_folder_hdfs = (
    "/path/to/company_data/sa/cpc/store/data/13_PROVINCES/" )

folder = [
    f.path for f in hdfs.get_file_info(fs.FileSelector(path_to_folder_hdfs)) ]

folder = sorted(folder)
folder
```

```python
df_profile = pd.read_parquet('/path/to/company_data/sa/cpc/store/data/PROFILE/df_profile.parquet',filesystem=hdfs)
df_profile= df_profile[['MA_DDO','TEN_DVIDCHINH_LV1','TEN_DVIDCHINH_LV2','TEN_DVIDCHINH_LV3','TEN_NN_LV1','TEN_NN_LV2','TEN_NN_LV3']]
df_profile
```

```python
dict_all_province = {}
for file in tqdm(folder):
    df_all_MDD = pd.read_parquet(
        file,
        filesystem=hdfs,)
    df_all_MDD = df_all_MDD[['MA_DDO','SLUONG_FINAL','Y','M']].merge(df_profile,on='MA_DDO',how='left')
    # lst = df_all_MDD.columns
    # detail_groupby = lst[5]
    # df_groupby = df_all_MDD.groupby(['Y','M',detail_groupby]).agg({lst[1]:'sum'}).reset_index()
    # df_groupby['date']= df_groupby.apply(lambda x:pd.to_datetime(f"{x.Y}-{x.M}-1"),axis=1)
    # df_total = df_groupby.pivot_table(index=detail_groupby,columns= 'date',values=lst[1]).T.fillna(0)
    dict_all_province[file] = df_all_MDD
```

```python
for file in tqdm(folder):
    df_all_MDD = dict_all_province[file]
    province = file.split('/')[-1].split('.')[0]
    df_all_MDD.to_parquet(f'/path/to/company_data/user/<user>/cpc/data/u2023/detail/no_breakdown/{province}.parquet',filesystem=hdfs,index=False)
```

```python
dict_df_total = {}
detail_groupby = 'DVDC_LV3'
detail_groupby1 = 'TEN_DVIDCHINH_LV1'
detail_groupby2 = 'TEN_DVIDCHINH_LV2'
detail_groupby3 = 'TEN_DVIDCHINH_LV3'
for file in tqdm(folder):
    df_MDD_province_copy = dict_all_province[file]
    df_MDD_province = df_MDD_province_copy.copy()
    df_MDD_province[detail_groupby1] = df_MDD_province[detail_groupby1].fillna('NONE')
    df_MDD_province[detail_groupby2] = df_MDD_province[detail_groupby2].fillna('NONE')
    df_MDD_province[detail_groupby3] = df_MDD_province[detail_groupby3].fillna('NONE')
    lst = df_MDD_province.columns
    df_groupby = df_MDD_province.groupby([detail_groupby1,detail_groupby2,detail_groupby3,'Y','M']).agg({lst[1]:'sum'}).reset_index()
    df_groupby['date']= df_groupby.apply(lambda x:pd.to_datetime(f"{x.Y}-{x.M}-1"),axis=1)
    df_total = df_groupby.pivot_table(index=[detail_groupby1,detail_groupby2, detail_groupby3],columns= 'date',values=lst[1]).T
    df_total.columns = df_total.columns.map(lambda x: f'{x[0]}-{x[1]}-{x[2]}')
    df_total_padding = pd.DataFrame({col: None for col in df_total.columns},
                                    index=pd.date_range(start='2023-10-01', end='2025-12-1', freq='MS'))
    df_total = pd.concat([df_total, df_total_padding], axis=0)
    mask = df_total.index.duplicated(keep='first')
    df_total = df_total[~mask]
    dict_df_total[file] = df_total
```

```python
for file in tqdm(folder):
    df = dict_df_total[file]
    province = file.split('/')[-1].split('.')[0]
    df.to_parquet(f'/path/to/company_data/user/<user>/cpc/data/u2023/detail/DVDC3/{province}.parquet',filesystem=hdfs)
```
